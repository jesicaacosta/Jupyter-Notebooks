### **Ejemplo Práctico de Análisis de Datos en Python**

```python
import pandas as pd

# Cargar datos simulados
data = {
    'Producto': ['A', 'B', 'C', 'D'],
    'Ventas': [100, 150, 80, 120],
    'Beneficio': [20, 30, 15, 25]
}
df = pd.DataFrame(data)

# Análisis básico: media, suma y correlación
print("Media de ventas:", df['Ventas'].mean())
print("Suma de beneficios:", df['Beneficio'].sum())
print("Correlación entre ventas y beneficios:")
print(df[['Ventas', 'Beneficio']].corr())

# Visualización simple
import matplotlib.pyplot as plt
df.plot(kind='bar', x='Producto', y='Ventas', title='Ventas por Producto')
plt.show()
```

### **Total de ventas en la semana**

```python
import pandas as pd
import matplotlib.pyplot as plt #Traigo libreria matploblib

# Cargar el archivo Excel
df = pd.read_excel('/content/ventas_supermercado.xlsx')

# Definir el orden de los días de la semana
dias_ordenados = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Asegurar que la columna 'Dia_Semana' esté en el orden correcto
df['Dia_Semana'] = pd.Categorical(df['Dia_Semana'], categories=dias_ordenados, ordered=True)

# Agrupar por día de la semana y calcular el total de ventas
ventas_por_dia = df.groupby('Dia_Semana')['Cantidad_Vendida'].sum()

# Graficar las ventas por día de la semana
ventas_por_dia.plot(kind='bar', color='skyblue')
plt.title('Total de Ventas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.show()
```

### **Identificación de tendencias en el comportamiento de los consumidores**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulación de datos de ventas en los últimos 12 meses
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
ventas = [1000, 1200, 1500, 1800, 2200, 2500, 2700, 2400, 2000, 2200, 2300, 2500]

# Crear DataFrame
df_ventas = pd.DataFrame({'Mes': meses, 'Ventas': ventas})

# Graficar la tendencia de ventas por mes
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', color='blue')
plt.title('Tendencia de Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### **Optimizacion en la cadena de suministro**

```python
import pandas as pd

# Simulación de tiempos de entrega de productos en días
datos_entregas = {'Producto': ['Producto A', 'Producto B', 'Producto C'],
                  'Tiempo_Entrega': [3, 5, 2],
                  'Costo_Entrega': [150, 200, 120]}

df_entregas = pd.DataFrame(datos_entregas)

# Identificación de productos con tiempos de entrega superiores a la media
media_tiempo_entrega = df_entregas['Tiempo_Entrega'].mean()
productos_lentos = df_entregas[df_entregas['Tiempo_Entrega'] > media_tiempo_entrega]

print(productos_lentos)
```

### **Análisis descriptivo de ventas**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulación de datos de ventas
datos_ventas = {'Producto': ['Producto A', 'Producto B', 'Producto C'],
                'Ventas': [100, 150, 120]}

df = pd.DataFrame(datos_ventas)

# Gráfico simple de ventas
df.plot(kind='bar', x='Producto', y='Ventas', color='lightblue')
plt.title('Análisis Descriptivo de Ventas')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Datos históricos de ventas y tiempo (meses)
tiempo = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
ventas = np.array([100, 120, 150, 170, 190, 210])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(tiempo, ventas)

# Predicción de las ventas para los próximos 3 meses
meses_futuros = np.array([7, 8, 9]).reshape(-1, 1)
prediccion_ventas = modelo.predict(meses_futuros)

print(f"Ventas previstas para los próximos meses: {prediccion_ventas}")
```

### **Visualización de Distribuciones**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de vehículos eléctricos
df = pd.read_excel('/content/Electric_Vehicle_Population_Data.xlsx')

# Establecer un estilo más claro para los gráficos
sns.set(style="whitegrid")

# Histograma para visualizar la distribución de la variable "Electric Range"
plt.figure(figsize=(10, 6))
sns.histplot(df['Electric Range'], kde=True, color='skyblue', bins=30)
plt.title('Distribución del Rango Eléctrico de Vehículos Eléctricos', fontsize=16)
plt.xlabel('Rango Eléctrico (millas)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Boxplot para visualizar el 'Base MSRP'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Base MSRP', data=df, color='lightgreen')
plt.title('Boxplot del Precio Base (MSRP)', fontsize=16)
plt.xlabel('Precio Base (USD)', fontsize=14)
plt.xticks(fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### **Análisis de Correlaciones**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de vehículos eléctricos
df = pd.read_excel('/content/Electric_Vehicle_Population_Data.xlsx')

# Filtrar solo las columnas numéricas
df_numerico = df.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlaciones
correlaciones = df_numerico.corr()

# Crear un heatmap para visualizar las correlaciones
plt.figure(figsize=(10, 8))  # Aumentar el tamaño del gráfico
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaciones entre Variables Numéricas', fontsize=16)
plt.show()
```

### **Visualización de Distribuciones Simuladas: Normal y Poisson**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos simulados de una distribución normal
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Visualizar la distribución normal
sns.histplot(data_normal, kde=True)
plt.title('Distribución Normal Simulada')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

# Generar datos simulados de una distribución de Poisson
data_poisson = np.random.poisson(lam=3, size=1000)

# Visualizar la distribución de Poisson
sns.histplot(data_poisson, kde=False)
plt.title('Distribución de Poisson Simulada')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

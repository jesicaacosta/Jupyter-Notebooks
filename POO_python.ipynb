```python
class Persona:
  def __init__(self,nombre,edad): #self es el primer par谩metro de los m茅todos de una clase y se usa para acceder a las variables y m茅todos de la instancia
    self.nombre = nombre
    self.edad = edad
```

```python
#Creo instancia de Persona

persona1 = Persona("Jesica", 30)

#Creo funcion para mostrar persona

def mostrar_info (n):
  print(f"Nombre: {n.nombre} y su edad es {n.edad}")

mostrar_info(persona1)
```

```python
persona2 = Persona("Jorge", 10)
persona3 = Persona("Maria", 20)
persona4 = Persona("Juan", 30)

personas = [persona2, persona3, persona4]

for p in personas:
  mostrar_info(p)
```

```python
class Animal: #clase base
  def __init__(self, nombre, raz):
    self.nombre = nombre
    self.raza =  raz

  def hacerSonido(self):
    print(f"{self.nombre} hace un sonido ")

  def mostrarPansa(self):
    print(f"{self.nombre} se tira de pansa")

#Creo clase que hereda de otra clase
class Perro(Animal): #hereda los atributos de la CLASE animal
  def hacerSonido(self):
    print(f"{self.nombre} esta ladrando ")

  def dormir(self):
    print(f"{self.nombre} esta durmiendo ")

  def mostrarPansa(self):
    print(f"{self.nombre} se tira de pansa porque es ")

perro1 = Perro("Kalel",'caniche')#CREO OBJ PERRO, QUE A SU VEZ HEREDA DE LA CLASE ANIMAL
perro1.hacerSonido()
perro1.mostrarPansa() #PUEDO O NO USAR TODOS LOS METODOS DE LAS CLASES
perro1.dormir() # METODO QUE SOLO ESTA EN LA CLASE PERRO


#Creo una clase nueva que hereda de otra clase
class Gatos(Animal):
  def hacerSonido(self):
    print(f"{self.nombre} esta maullando ")

  def mostrarPansa(self):
    print(f"{self.nombre} se tira de pansa porque es ")

#Creo instancia nueva que hereda de clase
gato1 = Gatos("Toto", 'gato')
gato1.hacerSonido()
gato1.mostrarPansa()


```

```python
class Perrita(Animal):
  def __init__(self, nombre,raz, c): #self, parametros de clase padre, parametros clase actual
    super().__init__(nombre, raz) #hereda clase padre
    self.color = c

pe1 = Perrita('Eileen',"Mestiza", "Negro")
print(f"{pe1.nombre} es de color {pe1.color} y su raza es {pe1.raza}.")

pe1.hacerSonido()
pe1.mostrarPansa()
```

```python
class Caminante:
    def __init__(self):
        print(f"Esta caminando")

class Koala (Animal, Caminante) : #hereda clase Animal y Caminante
  def __init__(self, nombre, raz):
    super().__init__(nombre, raz) #hereda todo de Animal
  def Trepa(self): #metodo propio de Koala
    print(f"Metodo propio: {self.nombre} esta trepando")

k = Koala("Koalita", "tre")
k.hacerSonido() #heredado de la clase Animal
k.mostrarPansa() #heredado de la clase Animal
k.Trepa() #heredado de la clase Caminante
```

**Encapsulacion y Variables privadas **

```python
class CuentaBancaria:
  def __init__(self,titular,saldo):
    self.titular = titular #atributo publico
    self.__saldo=saldo #atributo privado

  def mostrarSaldo(self):
    print(f"El saldo de {self.titular} es {self.__saldo}")

  def depositrar(self,monto):
    if monto > 0:
      self.__saldo += monto
      print(f"Se deposito {monto} a la cuenta de {self.titular}")
    else:
      print("El monto debe ser mayor a 0")

jesi = CuentaBancaria('Jesica Acosta', 100000)
jesi.mostrarSaldo()
jesi.depositrar(900)
jesi.mostrarSaldo()
print(jesi.titular)
#print(jesi.__saldo) # no puedo acceder directamente,
```

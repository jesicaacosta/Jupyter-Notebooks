```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Cargar los datos
df = pd.read_csv("/content/consumo_electrico_sectores_manufactureros_original.csv")

# Exploración inicial
print('Exploración inicial')
print(df.head())
print('describe: ',df.describe())

# Limpieza de datos (ejemplo: eliminar valores faltantes)
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['periodo'], df['molienda_cereales_y_oleaginosas'], marker='o')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

```python
# Gráfico de barras con Seaborn
sns.barplot(x='mes', y='ventas', data=df)
plt.title('Ventas Mensuales por Mes')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        'Col1': [100, 200, 300],
        'Col2': [400, 500, 600],
        'Col3': [700, 800, 900]
    },
    index = ['fila1', 'fila2', 'fila3']
)
df
```

üîç loc[]: Selecciona filas y columnas por etiquetas o nombres.

1.   Elemento de lista
2.   Elemento de lista



```python
df.loc['fila1']
```

```python
df.loc['fila1', 'Col2']
```

```python
type(df.loc['fila1'])
```

```python
df.loc[['fila1', 'fila2']] #localiza
```


El doble corchete [[]] en pandas se usa para seleccionar m√∫ltiples filas o columnas, permitiendo que el resultado sea un DataFrame y no solo una Serie.

```python
type(df.loc[['fila1', 'fila2']])
```

Segmentacion de filas

```python
df = pd.DataFrame(
    {
        'Col1': [100, 200, 300, 544],
        'Col2': [400, 500, 600, 28923],
        'Col3': [700, 800, 900, 343],
        'Col4': [4200, 5200, 620, 354],
        'Col5': [7009, 8090, 9090, 545]
    },index = ['fila1', 'fila2', 'fila3', 'fila4']
)
df.loc['fila2':'fila4']
```

Utilizar Booleanos para segmentar, en los que tebga True los va devolver

```python
df.loc[[True, False,False,False]]
```

Segmentar con condiciones

```python
df.loc[df['Col1']> 200]
```

TRAER COLUMNAS

```python
df.loc[:,[ 'Col1', 'Col2']] #del DF 'df' toma todas las filas : y √∫nicamente las columnas 'Col1' y 'Col2
```

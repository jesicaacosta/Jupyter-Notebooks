**✅ Obtener la media, mediana y rango del ranking global.**



El Global Rank en un dataset de universidades es la posición que ocupa cada universidad en un ranking mundial basado en ciertos criterios de evaluación. Es un número que indica qué tan bien posicionada está una universidad en comparación con todas las demás del mundo.



```python
#MEDIA GLOBAL
media_global = df_university['Global Rank'].mean()
media_global
```

```python
#MEDIANA
mediana_global = df_university['Global Rank'].median()
mediana_global
```

```python
#RANGO
rango_global_max= df_university['Global Rank'].max()
rango_global_max
rango_global_min = df_university['Global Rank'].min()
rango_global_min

print('Rango Global', rango_global_max, '-', rango_global_min )
```

✅ Identificar la universidad mejor rankeada en cada país.


```python
#Ordeno por Global Rank
mejor = df_university.sort_values('Global Rank')
mejor

mejor_rankeada = mejor.groupby('Country')['University']
mejor_rankeada = mejor_rankeada.first()
print('Universidad mejor rankeada en cada país')
mejor_rankeada.head(20)
```

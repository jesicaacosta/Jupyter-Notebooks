**MANEJO DE NULOS CON NUMPY**

```python
array_nulos = np.array([1,2,3,np.nan,5,np.nan, np.nan,7])
array_nulos
```

```python
#VERIFICA SI HAY NULOS EN EL ARRAY
np.isnan(array_nulos)
```

```python
np.isnan(array_concatenado).sum()
```

```python
#PROMEDIO DE ARRAY CON NULOS EN SU INTERIOR
np.mean(array_nulos)
```

```python
#PROMEDIO DE ARRAY CONSIDERANDO LS NULOS EN SU INTERIOR, NO CONSIDERA LOS NULOS LOS IGNORA
np.nanmean(array_nulos)
```

```python
#SUSTITUIR VALORES NULOS CON WHERE
array_con_0 = np.where(np.isnan(array_nulos),0,array_nulos)
#condicion, si se cumple reemplazo por esto, sino nos devuelve el valor en ese espacio del array
array_con_0
```

```python
promedio_array_con_0 = np.nanmean(array_nulos)
promedio_array_con_0
```

```python
#SUSTITUIR VALORES NULOS CON OTRA VARIABLE
array_con_variable = np.where(np.isnan(array_nulos),promedio_array_con_0 ,array_nulos)
#condicion, si se cumple reemplazo por esto, sino nos devuelve el valor en ese espacio del array
array_con_variable
```

```python
#array_filtrado = array_nulos[np.isnan(array_nulos)] me devuelve los nulos.
array_filtrado = array_nulos[~np.isnan(array_nulos)]  #me devuelve los no nulos
array_filtrado
# ~ es como un NOT
```

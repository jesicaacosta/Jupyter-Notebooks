```python
import funciones

#lista para guardar los datos de los alumnos
alumnos = []

print('Ingreso de alumnos: ')

#tomar los datos

for i in range(3):
         nombre = input(f'Ingrese el nombre del alumno {i +1}')
         edad = int(input(f'Ingrese la edad  del alumno'))
         dni = int(input(f'Ingrese el dni  del alumno'))
         gen = input(f'Ingrese el genero  del alumno')
         nota = input(f'Ingrese la nota  del alumno')


#agrego el alumno a la lista
alumnos.append({
        "nombre": nombre,
        "edad": edad,
        "dni": dni,
        "genero": gen.upper(),
        "nota": nota
    })


opc=input(print('Bienvenido al sistema, que desea hacer? '))
print('1. Listado de todos los alumnos')
print('2. El promedio de edades de los alumnos.')
print('3. La cantidad de alumnos que promocionaron (nota mayor o igual a 6).')
print('4. La cantidad de alumnos que aprobaron (nota mayor o igual a 4).')
print('5. La cantidad de alumnos que desaprobaron (nota menor a 4).')
print('6. El promedio de nota de los alumnos masculinos.')
print('7. El porcentaje de alumnas que promocionaron sobre el total de alumnas.')


# Menú de opciones
while True:
    print('\nBienvenido al sistema, ¿qué desea hacer?')
    print('1. Listado de todos los alumnos')
    print('2. Promedio de edades de los alumnos')
    print('3. Cantidad de alumnos que promocionaron (nota >= 6)')
    print('4. Cantidad de alumnos que aprobaron (nota >= 4)')
    print('5. Cantidad de alumnos que desaprobaron (nota < 4)')
    print('6. Promedio de nota de los alumnos masculinos')
    print('7. Porcentaje de alumnas que promocionaron sobre el total de alumnas')
    print('8. Salir')

    opc = input("Ingrese una opción: ")

    if opc == '1':
        listar_alumnos()
    elif opc == '2':
        promedio_edades()
    elif opc == '3':
        alumnos_promocionados()
    elif opc == '4':
        alumnos_aprobados()
    elif opc == '5':
        alumnos_desaprobados()
    elif opc == '6':
        promedio_notas_masculinos()
    elif opc == '7':
        porcentaje_promocionadas_femeninas()
    elif opc == '8':
        print("Saliendo del sistema...")
        break
    else:
        print("Opción no válida, por favor intente nuevamente.")
```

# Nueva sección

```python

# Definir funciones para las opciones del menú
def listar_alumnos():
    print("\nListado de todos los alumnos:")
    for alumno in alumnos:
        print(alumno)


def promedio_edades():
    promedio = sum(alumno["edad"] for alumno in alumnos) / len(alumnos)
    print(f"\nPromedio de edades de los alumnos: {promedio:.2f}")

def alumnos_promocionados():
    promocionados = sum(1 for alumno in alumnos if alumno["nota"] >= 6)
    print(f"\nCantidad de alumnos que promocionaron (nota >= 6): {promocionados}")

def alumnos_aprobados():
    aprobados = sum(1 for alumno in alumnos if alumno["nota"] >= 4)
    print(f"\nCantidad de alumnos que aprobaron (nota >= 4): {aprobados}")

def alumnos_desaprobados():
    desaprobados = sum(1 for alumno in alumnos if alumno["nota"] < 4)
    print(f"\nCantidad de alumnos que desaprobaron (nota < 4): {desaprobados}")

def promedio_notas_masculinos():
    notas_masculinos = [alumno["nota"] for alumno in alumnos if alumno["genero"] == 'M']
    if notas_masculinos:
        promedio = sum(notas_masculinos) / len(notas_masculinos)
        print(f"\nPromedio de notas de los alumnos masculinos: {promedio:.2f}")
    else:
        print("\nNo hay alumnos masculinos.")

def porcentaje_promocionadas_femeninas():
    alumnas = [alumno for alumno in alumnos if alumno["genero"] == 'F']
    if alumnas:
        promocionadas = sum(1 for alumna in alumnas if alumna["nota"] >= 6)
        porcentaje = (promocionadas / len(alumnas)) * 100
        print(f"\nPorcentaje de alumnas que promocionaron: {porcentaje:.2f}%")
    else:
        print("\nNo hay alumnas.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("//content//drive//MyDrive//CURSO IA//Analisis Exploratorio y visualizaciON//ventas_supermercado.xlsx")
df.head()
```

```python
#ordenar
dias_ordenados = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
```

```python
#ordenar columna dia de semana como dias_ordenados
df['Dia_Semana'] = pd.Categorical(df['Dia_Semana'], categories=dias_ordenados, ordered=True)
#Categorical ees un tipo de dato, q representa datos que tienen un número fijo de valores posibles (categorías)
```

```python
#AGRUPAR POR DIA DE LA SEMA Y CALCULAR EL TOTAL DE VENTAS

ventas_por_dia = df.groupby('Dia_Semana')['Cantidad_Vendida'].sum()
ventas_por_dia
```

```python
#GRAFICO DE BARRAS X,Y
#ventas_por_dia = df.groupby('Dia_Semana')['Cantidad_Vendida'].sum() --> x=Dia_Semana, y= CantidadVendida
ventas_por_dia.plot(kind= 'bar', color= 'pink') #ventasxdia es la var q guarda los datos agrupados
#.plot crea grafico a partir de datos en objetos pandas
plt.title('Total de ventas por día de la semana') #titulo del grafico
plt.xlabel('Dia de la semana')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45,color='blue')
plt.show()
```

```python
date_tt = pd.read_csv("//content//drive//MyDrive//DATASETS//TickTok Dataset 2.csv", encoding='latin-1')
date_tt.head()
```

```python
#cambiar nombres de columnas a castellano
#creo una nueva lista de nombres
nuevos = { 'Celebrity' :'Celebridad',
          'Followers'   : 'Seguidos',
           'Following': 'Seguidores',
           'Likes': 'Me Gusta',
           'T.Videos'   : 'Videos Totales',
           'Video Duration': 'Duracion_Videos',
          'Average Views': 'Promedio_Vistas',
           'Net Worth(Million)': 'Patrimonio_Neto',
           'Most Viewed Video(Billion)' : 'Video_Mas_Visto(B)',
           'Most Liked Video(Million)': 'Video_Mas_Likeado(M)',
           'Video Category': 'Categoria_Video'
           }
date_tt.rename(columns=nuevos, inplace=True)
date_tt.head()
```

```python
mas_like = date_tt.groupby('Celebridad')['Video_Mas_Likeado(M)'].sum()
mas_like.sort_values(ascending=False).head(5)
```

```python
mas_like.plot(kind='bar', color= 'pink')
plt.title('Mas Likeados') #titulo del grafico
plt.xlabel('Celebridad')
plt.ylabel('Video_Mas_Likeado')
plt.xticks(rotation=45,color='blue')
plt.show()
```

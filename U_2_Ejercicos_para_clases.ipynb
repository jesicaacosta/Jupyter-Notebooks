```python
# BLOQUE 1: Operaciones básicas con Listas
# Crear una lista y realizar operaciones
print("BLOQUE 1: Operaciones con Listas")
frutas = ["manzana", "naranja", "pera", "uva"]
print("Lista inicial:", frutas)

# Añadir un elemento al final de la lista
frutas.append("mango")
print("Lista después de añadir un mango:", frutas)

# Eliminar un elemento de la lista
frutas.remove("pera")
print("Lista después de eliminar una pera:", frutas)

# Contar cuántas veces aparece "naranja" en la lista
print("Número de naranjas:", frutas.count("naranja"))

# Invertir la lista
frutas.reverse()
print("Lista invertida:", frutas)
```

# **Bloque 2: Operaciones con tuplas (desempaquetado, inmutabilidad).**

```python
# BLOQUE 2: Operaciones con Tuplas
print("\nBLOQUE 2: Operaciones con Tuplas")
tupla_frutas = ("manzana", "naranja", "uva")
print("Tupla inicial:", tupla_frutas)

# Intentar modificar un valor de la tupla (esto generará un error)
try: #intentar hacer
    tupla_frutas[0] = "pera"
except TypeError as e: #si hay un error: 'as e' asignar el msj de error a e
    print("Error: No se puede modificar una tupla. Detalle:", e)

# Desempaquetar una tupla
fruta1, fruta2, fruta3 = tupla_frutas
print('Tupla entera: ', tupla_frutas)
print('Tupla desempaquetada: ')
print("1:", fruta1, "2:", fruta2, "3:", fruta3)
```

# **Bloque 3: Operaciones con diccionarios (agregar, eliminar, acceder).**

```python
# BLOQUE 3: Operaciones con Diccionarios
print("\nBLOQUE 3: Operaciones con Diccionarios")
estudiante = {
    "nombre": "Ana",
    "edad": 22,
    "carrera": "Biología"
}
print("Diccionario inicial:", estudiante)

# Añadir una nueva clave y valor al diccionario
estudiante["grado"] = "Tercer Año"
print("Diccionario después de añadir un grado:", estudiante)

# Eliminar una clave
del estudiante["carrera"]
print("Diccionario después de eliminar la carrera:", estudiante)

# Acceder a un valor usando una clave
print("Nombre del estudiante:", estudiante["nombre"])
```

# **Bloque 4: Operaciones con NumPy (creación de arrays, productos escalares, transposición, norma).**

```python
# BLOQUE 4: Operaciones con NumPy
import numpy as np
print("\nBLOQUE 4: Operaciones con NumPy")

# Crear un array NumPy de una dimensión
array_1D = np.array([1, 2, 3, 4, 5])
print("Array de una dimensión:", array_1D)

# Crear una matriz (array de 2 dimensiones)
matriz = np.array([[1, 2], [3, 4]])
print("Matriz de 2x2:\n", matriz)

# Producto escalar entre dos vectores
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
producto_escalar = np.dot(vector1, vector2)
print("Producto escalar de los vectores:", producto_escalar)

# Transponer una matriz
matriz_transpuesta = np.transpose(matriz)
print("Matriz transpuesta:\n", matriz_transpuesta)

# Norma de un vector
norma_vector = np.linalg.norm(vector1)
print("Norma del vector1:", norma_vector)
```

# **Bloque 5: Ejemplos avanzados con diccionarios y tuplas (diccionarios con tuplas como claves).**

```python

# BLOQUE 5: Ejemplos avanzados con Diccionarios y Tuplas
print("\nBLOQUE 5: Ejemplos avanzados con Diccionarios y Tuplas")

# Crear un diccionario que contenga tuplas como claves y listas como valores
diccionario_avanzado = {
    (1, 2): ["A", "B"],
    (3, 4): ["C", "D"]
}
print("Diccionario con tuplas como claves:", diccionario_avanzado)

# Acceder a un valor en el diccionario
clave = (1, 2)
print("Valor asociado a la clave (1, 2):", diccionario_avanzado[clave])

# Iterar sobre un diccionario con tuplas como claves
for key, value in diccionario_avanzado.items():
    print(f"Clave: {key}, Valor: {value}")
```

# **Bloque 6: Creación de diccionarios y conversión de tipos de datos.**

```python
 #BLOQUE 6: Creación de un diccionario usando dict() y conversión de tipos de datos
print("\nBLOQUE 6: Creación y conversión de tipos de datos en Diccionarios")
persona = dict(
    nombre="Guido",
    apellido="van Rossum",
    trabajo="Creador de Python"
)
print("Diccionario creado con dict():", persona)

# Convertir una lista de tuplas en un diccionario
lista_de_tuplas = [("nombre", "Ana"), ("edad", 25), ("carrera", "Ingeniería")]
diccionario_convertido = dict(lista_de_tuplas)
print("Diccionario convertido desde lista de tuplas:", diccionario_convertido)
```

# **Bloque 7: Operaciones avanzadas con matrices en NumPy (producto de matrices, norma)**

```python
# BLOQUE 7: Más operaciones con NumPy - Producto de matrices y norma
print("\nBLOQUE 7: Producto de matrices y norma con NumPy")

# Crear dos matrices compatibles para multiplicación
matriz_a = np.array([[1, 2], [3, 4]])
matriz_b = np.array([[5, 6], [7, 8]])

# Producto matricial
producto_matricial = np.dot(matriz_a, matriz_b)
print("Producto matricial:\n", producto_matricial)

# Norma de una matriz
norma_matriz = np.linalg.norm(matriz_a)
print("Norma de la matriz A:", norma_matriz)

```

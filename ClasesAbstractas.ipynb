```python
from abc import ABC, abstractmethod
"""ABC es la clase base para definir clases abstractas. Al usarla,
 estás indicando que no se puede instanciar directamente y que puede contener métodos abstractos."""

class Vehiculo(ABC):
  #Cuando una clase hereda de ABC (que proviene del módulo abc), se convierte en una clase abstracta.
  @abstractmethod
  def ir(self): #debe ser implementado por cualquier subclase que herede de Vehiculo
    #pass #pass omite
    print("El vehiculo va")
  def frenar(self):  #puede ser usado por las subclases sin necesidad de implementarl
    print("METODO DEL PADREa")

class Coche(Vehiculo):
  def ir(self):
    print("El coche va")

class Bicicleta(Vehiculo):
  def ir(self):
    print("La bicicleta va")
    def frenar(self):
      print("El Bicicleta frena")
```

```python
veh = Coche()
veh.ir()
veh.frenar()
bici = Bicicleta()
bici.ir()
#veh = Vehiculo() #TypeError: Can't instantiate abstract class Vehiculo with abstract method ir
```

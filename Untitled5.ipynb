```python
import plotly.express as px
import pandas as pd

# Datos de muestra
df = pd.DataFrame(dict(
    grupo = ["A", "B", "C", "D", "E"],
    valor = [14, 12, 8, 10, 16]))
```

```python
fig = px.bar(df, x = 'grupo', y = 'valor', color = 'grupo',
             color_discrete_map = {'A': '#7FD4C1', 'B': '#30BFDD', 'C': '#8690FF',
                                   'D': '#ACD0F4', 'E': '#F7C0BB'})

fig.show()
```
